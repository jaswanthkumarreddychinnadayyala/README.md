```python
import pandas as pd

# Load the Tata Steel dataset
df = pd.read_csv('/content/tatasteel_2015_2025 (1).csv')

# Show first few rows
df.head()
```

```python
# Check structure and data types
print(df.info())

# Check for missing/null values
print("\nMissing values:\n", df.isnull().sum())
```

```python
# Show the column names
print(df.columns)

# Show the index type
print("\nIndex type:", type(df.index))

# Preview the first few rows
df.head()
```

```python
# Create a new 'Price' column as the closing price
df['Price'] = df['Close']
```

```python
# Add static company-level information
df['P/E Ratio'] = 58.08
df['Face Value'] = 1
df['Market Cap'] = '1.86L Cr'         # Example value
df['Div Yield'] = '3.2%'
df['52-wk High'] = 171.3
df['52-wk Low'] = 103.1
```

```python
df.dtypes
```

```python
# Convert 'Price' column to numeric (coerce = force errors to NaN)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

```python
# Moving Averages
df['SMA_10'] = df['Price'].rolling(window=10).mean()
df['SMA_50'] = df['Price'].rolling(window=50).mean()

# Daily Return
df['Daily_Return'] = df['Price'].pct_change()

# Volatility
df['Volatility_10'] = df['Daily_Return'].rolling(window=10).std()

# EMA
df['EMA_20'] = df['Price'].ewm(span=20, adjust=False).mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
# Drop rows with any NaN (from moving averages, returns, etc.)
df = df.dropna()

# Features (X) and Target (y)
features = ['SMA_10', 'SMA_50', 'EMA_20', 'Daily_Return', 'Volatility_10']
X = df[features]
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(y_test.index, y_test.values, label='Actual Price', color='blue')
plt.plot(y_test.index, y_pred, label='Predicted Price', color='orange')
plt.title('Tata Steel: Actual vs Predicted Prices')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```
